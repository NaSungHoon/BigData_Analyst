```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as ec
import re

path = r'C:\Users\medici\Downloads\driver\chromedriver.exe'
s = Service(path)
driver = webdriver.Chrome(service = s)

url = f"https://sports.news.naver.com/kbaseball/schedule/index?date=20230401&month=03&year=2022&teamCode="
driver.get(url)
driver.implicitly_wait(2) 
driver.set_window_size(1000,800)

a= []
for _ in range(8):
    driver.find_element(By.XPATH,'//*[@id="content"]/div[1]/div[2]/div[1]/button[2]').click()
    for k in range(1,len(driver.find_element(By.XPATH,'//*[@id="calendarWrap"]').find_elements(By.TAG_NAME,'div'))+1):
        if driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr/td[3]/span').text == '프로야구 경기가 없습니다.':
            continue
        else:    
            for i in range(1,len(driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody').find_elements(By.TAG_NAME,'tr'))+1):
                if driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr['+str(i)+']').find_elements(By.CLASS_NAME,'add_state') ==[]:
                    if len(driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr['+str(i)+']').find_elements(By.TAG_NAME,'td'))==7:
                        c2=driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr['+str(i)+']/td[3]/span[1]').text
                        temp = driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr['+str(i)+']/td[3]/strong').text
                        c3=temp[0]
                        c4=temp[2]
                        c5=driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr[1]/td[1]/span/strong').text
                    else:
                        c2=driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr['+str(i)+']/td[2]/span[1]').text
                        temp=driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr['+str(i)+']/td[2]/strong').text
                        c3=temp[0]
                        c4=temp[2]
                    c5=driver.find_element(By.XPATH,'//*[@id="calendarWrap"]/div['+str(k)+']/table/tbody/tr[1]/td[1]/span/strong').text
                    a.append([c2,c3,c4,c5])
                else:
                    pass
    
```
