```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as ec
import re

########
from collections import deque
#######



path = r'C:\Users\medici\Downloads\driver\chromedriver.exe'
s = Service(path)
driver = webdriver.Chrome(service = s)

url = f"https://strikes.zone/game/221108KWHSSG"
driver.get(url)
driver.implicitly_wait(2) 
driver.set_window_size(1000,800)

a= deque([])


for x in range(4,12):
    t1 = time.time()
    driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[1]/input').click()
    time.sleep(2)               
    driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[1]/a[6]').click()
    time.sleep(2)                
    driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[2]/div[1]/span[3]').click()
    time.sleep(2)                  
    driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[2]/div[2]/span['+str(x)+']').click()
    time.sleep(2)      
    for w in range(42):
        
        if driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[2]/table/tbody').find_elements(By.TAG_NAME,'td')[w].get_attribute('class')=='cell cur-month':
            driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[2]/table/tbody').find_elements(By.TAG_NAME,'td')[w].click()
            time.sleep(1)
            driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[2]/button').click()
            time.sleep(1)
        else:
            continue
    
        for k in range(1,len(driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/ul').find_elements(By.TAG_NAME,'li'))+1):    
            driver.find_element(By.XPATH,'/html/body/div/div/div/div/section/aside/ul/li['+str(k)+']').click()
            time.sleep(3)  
            for i in range(len(driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/section/section/div/div[3]/div[1]/div/div').find_elements(By.TAG_NAME,'image'))):
                c_0 = driver.find_element(By.XPATH,"//*[@id='__layout']/div/div/section/section/section/div/div[3]/div[1]/div/div/*[name()='svg']").find_elements(By.TAG_NAME,'image')[i].get_attribute('x')
                c_1 = driver.find_element(By.XPATH,"//*[@id='__layout']/div/div/section/section/section/div/div[3]/div[1]/div/div/*[name()='svg']").find_elements(By.TAG_NAME,'image')[i].get_attribute('y')
                if driver.find_element(By.XPATH,"//*[@id='__layout']/div/div/section/section/section/div/div[3]/div[1]/div/div/*[name()='svg']").find_elements(By.TAG_NAME,'image')[i].get_attribute('xlink:href') =='/_nuxt/img/1c8f58b.png':
                    c_2 = 'ball'
                else:
                    c_2 = 'strike'  
                c_3 = driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/section/section/div/div[3]/div[1]/div/header').text.replace(' 공격','')
                c_4 = '공격'
                c_5 = re.findall(r'\d{4}.\d{2}.\d{2}',driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/section/header/div[1]/div[1]').text.replace(' ',''))[0]
                if re.findall(r'\(\w{5}\)',driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/section/header/div[1]/div[1]').text.replace(' ','')) == []:
                    c_6 = ''
                else:    
                    c_6 = re.sub(r'\(|\)','',re.findall(r'\(\w{5}\)',driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/section/header/div[1]/div[1]').text.replace(' ',''))[0])
                a.append([c_0,c_1,c_2,c_3,c_4,c_5,c_6])

            for i in range(len(driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/section/section/div/div[3]/div[2]/div/div').find_elements(By.TAG_NAME,'image'))):
                b_0 = driver.find_element(By.XPATH,"//*[@id='__layout']/div/div/section/section/section/div/div[3]/div[2]/div/div/*[name()='svg']").find_elements(By.TAG_NAME,'image')[i].get_attribute('x')
                b_1 = driver.find_element(By.XPATH,"//*[@id='__layout']/div/div/section/section/section/div/div[3]/div[2]/div/div/*[name()='svg']").find_elements(By.TAG_NAME,'image')[i].get_attribute('y')
                if driver.find_element(By.XPATH,"//*[@id='__layout']/div/div/section/section/section/div/div[3]/div[2]/div/div/*[name()='svg']").find_elements(By.TAG_NAME,'image')[i].get_attribute('xlink:href') =='/_nuxt/img/1c8f58b.png':
                    b_2 = 'ball'
                else:
                    b_2 = 'strike'    
                b_4 = '수비'               
                a.append([b_0,b_1,b_2,c_3,b_4,c_5,c_6])
        driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[1]/input').click()
        time.sleep(1)
        driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[1]/a[6]').click()
        time.sleep(1)                  
        driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[2]/div[1]/span[3]').click()
        time.sleep(1)                
        driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section/aside/div/span[2]/div/div[2]/div[1]/div[2]/div[2]/span['+str(x)+']').click()
        time.sleep(1)
```
